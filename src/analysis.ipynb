```python
import pandas as pd
import glob
import os
import matplotlib.pyplot as plt
%matplotlib inline
```

## Start from one data file...

```python
data = pd.read_csv('../exampleco_data/machine_0.csv')
```

```python
def plot_histgram(data):
    fig, (ax1, ax2, ax3, ax4) = plt.subplots(1, 4, figsize = (20, 5))
    fig.suptitle('Histogram plot of each metric value')
    ax1.hist(data['0'], bins = 20)
    ax1.set(title='sensor_0',xlabel='value',ylabel='frequency')
    ax2.hist(data['1'], bins = 20)
    ax2.set(title='sensor_1',xlabel='value',ylabel='frequency')
    ax3.hist(data['2'], bins = 20)
    ax3.set(title='sensor_2',xlabel='value',ylabel='frequency')
    ax4.hist(data['3'], bins = 20)
    ax4.set(title='sensor_3',xlabel='value',ylabel='frequency')
    plt.show()
```

```python
def plot_scatter(data):
    fig, (ax1, ax2, ax3, ax4) = plt.subplots(1, 4, figsize = (20, 5))
    fig.suptitle('Scatter plot of each metric value')
    ax1.scatter(x=range(len(data)), y=data['0'])
    ax1.set(title='sensor_0', ylabel='value')
    ax2.scatter(x=range(len(data)), y=data['1'])
    ax2.set(title='sensor_1', ylabel='value')
    ax3.scatter(x=range(len(data)), y=data['2'])
    ax3.set(title='sensor_2', ylabel='value')
    ax4.scatter(x=range(len(data)), y=data['3'])
    ax4.set(title='sensor_3', ylabel='value')
    plt.show()
```

## Plot the data to understand the distribution of the metric values

```python
plot_histgram(data)
```

As the histograms shown above, all four metric values center around 0, which indicate the machine failed after a short period of normal functioning. There are also some outliers that fall into the ranges of <-200 and >200.

```python
plot_scatter(data)
```

The scatter plots give a better picture of the machine status over the time and the distribution of the outliers.

## Extend analysis to all data files...

```python
# read all csv files in the directory
all_files = glob.glob('../exampleco_data/*.csv')
for file in all_files:
    df = pd.read_csv(file)
    print(os.path.basename(file).split('.')[0])
    plot_scatter(df)
```

Each machine has slightly different patterns but they all have obvious outliers that need to be removed.
